```python
#Method #2: Mean Pixel Value of Channels
```

```python
dataset_path = '../input'
img2 = imread(os.path.join(dataset_path, 'woman-and-girl-painting-3171120.jpg'))
img2.shape

```

```python
feature_matrix = np.zeros((958,640)) 
feature_matrix.shape
```

```python
for i in range(0,img2.shape[0]):
    for j in range(0,img2.shape[1]):
        feature_matrix[i][j] = ((int(img2[i,j,0]) + int(img2[i,j,1]) + int(img2[i,j,2]))/3)
```

```python
features = np.reshape(feature_matrix, (958*640)) 
features.shape
```

```python
#Method #3: Extracting Edge Features
```

```python

#importing the required libraries
import numpy as np
from skimage.io import imread, imshow
from skimage.filters import prewitt_h,prewitt_v
import matplotlib.pyplot as plt
%matplotlib inline


#calculating horizontal edges using prewitt kernel
edges_prewitt_horizontal = prewitt_h(img_gray)
#calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(img_gray)

imshow(edges_prewitt_vertical, cmap='gray')
```
